```python
import chromadb

client = chromadb.Client()
collection = client.get_collection(name="my_collection")
```

```python
collection.add(
    documents=["This is kfc chicken."],
    documents=["This is egusi soup."],
    ids=["id1", "id2"]
)
```

```python
all_doc = collection.get()
all_doc
```

```python
documents = collection.get(
    ids=['id1']
)
```

```python
results = collection.query(
    query_texts=["This is a query about Nigerian "],
    n_results=2
)

results
```

```python
collection.delete(ids=all_doc["ids"])
collection.get()
```
