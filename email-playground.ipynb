```python
from langchain_groq import ChatGroq
from dotenv import load_dotenv

import os
load_dotenv()

model = ChatGroq(
    model = "llama-3.1-8b-instant",
    groq_api_key = os.getenv('GROQ_API_KEY'),
    temperature = 0
)

response = model.invoke('fastest route to get hired as an agentic ai engineer')
print(response.content)
```

```python
from langchain_community.document_loaders import WebBaseLoader
loader =  WebBaseLoader("https://www.linkedin.com/jobs/view/4415145028/")
data = loader.load().pop().page_content
print(data)
```

```python
from langchain_core.prompts import PromptTemplate

prompt_extract = PromptTemplate.from_template(
    """
    ### SCRAPED TEXT FROM A WEBSITE:
    {data}
    ### INSTRUCTIONS:
    This is a scaped text from a career's webpage'
    Your job is to extract the job description and return them in json format containing this keys: 'industry','role', 'experience', 'skills', 'description'.
    return only valid json
    ### NO PREAMBLE
    """
)

chain_extract = prompt_extract | model
res = chain_extract.invoke(input={'data':data})
print(res.content)
```

```python
from langchain_core.output_parsers import JsonOutputParser
jsonparser = JsonOutputParser()
json_res = jsonparser.parse(res.content)
json_res
```

```python
import chromadb
import uuid

client = chromadb.PersistentClient('vector_store')
collections = client.get_or_create_collection(name = "my_portfolio")

for _,row in df.iterrows():
    if not collections.count():
        collections.add(
            documents= row['Techstack'],
            metadatas={'links': row['Links']},
            ids=[str(uuid.uuid4())]
        )

client = chromadb.PersistentClient('vector_store')
collections = client.get_or_create_collection(name = "my_portfolio")

for _,row in df.iterrows():
    if not collections.count():
        collections.add(
            documents= row['Industry'],
            metadatas={'links': row['Links']},
            ids=[str(uuid.uuid4())]
        )
```

```python
job = json_res
links = collections.query(
    query_texts = job['skills'],
    n_results=2
) .get('metadatas')
links
```

```python
prompt_industry = PromptTemplate.from_template(
"""
### JOB DESCRIPTION / COMPANY INFO:
{job_description}

### COMPANY NAME (if known):
{company}

### INSTRUCTION:
Based on the job description and company name above, identify the single industry this company
most likely operates in (e.g. Retail, Healthcare, Finance, Banking, Insurance, Education,
Real Estate, Logistics, Manufacturing, Travel and Hospitality, Media and Entertainment,
Telecommunications, Energy and Utilities, Agriculture, Government and Public Sector,
Sports and Fitness, Legal, Automotive, Non-profit, E-commerce, or another industry if none of
these fit).
Return ONLY the industry name as a plain string, nothing else - no preamble, no punctuation.
### INDUSTRY:
"""
)

chain_industry = prompt_industry | model
job = json_res
industry_res = chain_industry.invoke({
    "job_description": job.get("description", ""),
    "company": job.get("company", "")
})
client_industry = industry_res.content.strip()
client_industry
```

```python
industry_links = collections.query(
    query_texts = job['industry'],
    n_results=2
) .get('metadatas')
industry_links
```

```python
prompt_email = PromptTemplate.from_template(
"""
### JOB DESCRIPTION:
{job_description}

### CLIENT INDUSTRY:
{client_industry}

### INSTRUCTION:
You are Dominion, an AI Engineer seeking a Software Engineer role. You are highly skilled in software development, distributed systems, cloud-native architectures, and modern data and AI systems (Python, SQL, Spark, Databricks, AWS/GCP/Azure, APIs, and agentic AI frameworks).
Your task is to write a short, compelling cold email to the hiring team based on the job description above, showing how your skills align with their needs and how you can contribute to building scalable software and intelligent data systems.
The company operates in the {client_industry} industry - briefly show awareness of that domain's needs/challenges where relevant.
Also include relevant project/portfolio links from the following two lists to strengthen credibility:
- Tech-stack portfolio links (relevant technical work): {tech_link_list}
- Industry portfolio links (relevant domain experience in {client_industry}): {industry_link_list}
Remember, you are Dominion applying for a Software Engineer role.
Do not add a preamble.
### EMAIL (NO PREAMBLE):
"""
)

chain_email = prompt_email | model
res = chain_email.invoke({
    "job_description": str(job),
    "client_industry": client_industry,
    "tech_link_list": links,
    "industry_link_list": industry_links
})
print(res.content)
```
